```python
from bing_image_downloader import downloader
import shutil
from pathlib import Path
import os

def download_images(keyword, num_images, file_name):
    output_directory = "images"

    # 출력 디렉토리 생성
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # 이미지 다운로드 시도
    downloader.download(keyword, limit=num_images, output_dir=output_directory,
                        adult_filter_off=True, force_replace=False, timeout=60)

    root = Path().cwd() / output_directory / keyword  # keyword를 포함한 디렉토리로 변경
    downloaded_files = list(root.glob("*.*"))

    print(f"다운로드된 파일 경로: {root}")
    print(f"다운로드된 파일 목록: {downloaded_files}")

    if not downloaded_files:
        print("이미지가 다운로드되지 않았습니다. 검색어와 인터넷 연결을 확인하세요.")
        return

    # 다운로드된 파일 이름 변경
    for i, file in enumerate(downloaded_files):
        extension = file.suffix
        new_file_name = f"{file_name}{i+1}{extension}"
        new_file_path = root / new_file_name
        file.rename(new_file_path)

    print(f"{len(downloaded_files)}개의 이미지가 다운로드되었습니다.")

# 이미지 다운로드를 위한 변수 설정 및 함수 호출
keyword = "plasticbottle"
num_images = 100
file_name = "plastic_bottle"

# 함수 호출
download_images(keyword, num_images, file_name)
```

```python
# 폴더를 압축하는 함수 정의
def compress_folder(folder_path, output_filename):
    shutil.make_archive(output_filename, 'zip', folder_path)

folder_path = "images"
output_filename = "images_archive"
compress_folder(folder_path, output_filename)
```

```python
# 폴더 내의 파일 목록 출력
root = Path(folder_path)
if not root.exists():
    root.mkdir(parents=True, exist_ok=True)

print("폴더 내의 파일 목록:")
for file in root.iterdir():
    if file.is_file():
        print(file.name)
```
