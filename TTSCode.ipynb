```python
text1 = "안녕하세요, 페트병 분리수거를 시작해 볼까요?"
text2 = "사용 방법을 알려드리겠습니다."
text3 = "먼저, 적립하기 버튼을 눌러주세요"
text4 = "포인트를 적립할 본인 휴대전화 번호를 입력해 주세요."
text5 = "페트병을 투입한 후, 사진을 찍어주시면, 분리수거 가능 여부 확인이 가능합니다."
text6 = "투입이 가능하다면, 적립 완료 버튼을 클릭하면 됩니다."
text7 = "마지막 화면에서 오늘 적립된 포인트와 총 누적 포인트를 확인할 수 있습니다."
```

```python
from gtts import gTTS
from IPython.display import Audio, display

texts = [text1, text2, text3, text4, text5, text6, text7]
for i, text in enumerate(texts):
  tts = gTTS(text = text, lang = 'ko')
  filename = f"output_{i+1}.mp3"
  tts.save(filename)
  display(Audio(filename, rate=22050))
```
